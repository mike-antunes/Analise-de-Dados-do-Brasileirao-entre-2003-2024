# Dados das 21 Temporadas do Brasileirão Entre (2003 até 2024)

```python
import pandas as pd
```

```python
df = pd.read_csv('brasileirao.csv')
df
```

- Susbtituindo os nomes das colunas para facilitar a análise

```python
df = df.rename(columns={"season": "temporada", "place": "posição", "team": "time", "points": "pontos", "played": "jogos", "won": "vitorias",
                        "draw": "empates", "loss": "derrotas", "goals": "gols_marcados", "goals_taken": "gols_sofridos",
                          "goals_diff": "saldo_de_gols"})

df
```

```python
df.info()
```

- Vamos adicionar novas colunas ao Dataset nos retornando a porcentagem estatística do desempenho dos times em cada temporada

```python
df["percentual_vitorias"]= df["vitorias"] / df["jogos"] * 100
df["percentual_empates"]= df["empates"] / df["jogos"] * 100
df["percentual_derrotas"]= df["derrotas"] / df["jogos"] * 100
df["media_gols_marcados"]= df["gols_marcados"] / df["jogos"]
df["media_gols_sofridos"]= df["gols_sofridos"] / df["jogos"]

df
```

- **Mas afinal, como vamos medir a correlação para saber o que mais pesa em um critério para tornar um time campeão ou rebaixado????**
- **Quais parâmentros são determinantes para atingir tais resultados??**

## Construindo correlações para descobrir quais critérios são importantes para definir a classificação de um time

- Vamos descobrir utilizando **.corr** a importância de cada coluna que leva um time a ser campeão e rebaixado

```python
# ANALISANDO CRITÉRIOS MAIS IMPORTANTES PARA SER CAMPEÃO DO BRASILEIRÃO

df["campeao"] = (df["posição"] == 1)

df["rebaixado"] = (df["posição"] >= 17)

cor_campeao = (
    df.corr(numeric_only=True)["campeao"]
    .sort_values(key=abs, ascending=False)
)

print(cor_campeao)


#NOTE QUE = Os critérios mais importantes para ser campeão do Brasileirão é o percentual de vitórias, seguido pelo número de vitórias,
#número de pontos, saldo de gols e assim sucessivamente
```

```python
# ANALISANDO CRITÉRIOS MAIS IMPORTANTES PARA SER UM TIME REBAIXADO NO BRASILEIRÃO


cor_rebaixado = (
    df.corr(numeric_only=True)["rebaixado"]
    .sort_values(key=abs, ascending=False)
)

print(cor_rebaixado)


#NOTE QUE = No caso dos times rebaixados, os critérios mais importantes foram sua posição obviamente definindo seu rebaixamento, seguido
#pelo número de derrotas, saldo de gols, percentual de derrotas, percentual de vitórias e assim sucessivamente em ordem decrescente
```

- **OBSERVAÇÃO SOBRE O QUE VIMOS =** Podemos dizer que os dados de metricas positivas se enquadram aos times campeões, enquanto as metricas
negativas sempre se enquadram aos times rebaixados.

- Utilizando **Feature Importance** com **Random Forest** para nos retornar as métricas mais importantes para explicar as posições dos times

```python
#PACOTE UTILIZADO PARA CRIAR O MODELO DE MACHINE LEARNING
from sklearn.ensemble import RandomForestClassifier


#LISTA DE VARIÁVEIS UTILIZADAS
X = df[["vitorias", "percentual_vitorias", "derrotas", "percentual_derrotas", "saldo_de_gols", "gols_marcados", "gols_sofridos", 
        "pontos", "media_gols_marcados", "media_gols_sofridos", ]]


#VARIÁVEL ALVO
y = (df["posição"] == 1)


#TREINAMENTO DO MODELO
modelo = RandomForestClassifier(
    n_estimators=500,
    random_state=42
)

modelo.fit(X, y)


#A IMPORTÂNCIA DAS VARIÁVEIS
importancias = pd.DataFrame({
    "variavel": X.columns,
    "importancia": modelo.feature_importances_
})

importancias.sort_values(
    by="importancia",
    ascending=False
)


#NOTE QUE= Em primeiro lugar para descrever a posição do time segundo nosso modelo foi o **percentual de vitórias**.
#Sendo assim, quanto mais um time ganhar suas partidas, melhor será sua posição na tabela do Brasileirão.
```

```python
# Gráfico
import matplotlib.pyplot as plt

importancias.sort_values(
    by="importancia",
    ascending=True
).plot(
    x="variavel",
    y="importancia",
    kind="barh",
    figsize=(10, 6),
    legend=False,
    title="Importância das Variáveis"
)

plt.xlabel("Importância")
plt.ylabel("Variável")
plt.tight_layout()
plt.show()
```

- **Afinal, o que é mais importante para um time ser campeão? Um ataque marcando muitos gols ou uma defesa forte tomando poucos gols????**

- Vamos comparar o peso de marcar mais gols e sofrer menos gols para justificar as posições dos times utilizando novamente **Random Forest**

```python
X = df[["gols_marcados", "gols_sofridos"]]

y = (df["posição"] == 1)

modelo = RandomForestClassifier(
    n_estimators=500,
    random_state=42
)
modelo.fit(X, y)


importancias = pd.DataFrame({
    "variavel": X.columns,
    "importancia": modelo.feature_importances_
})

print(importancias.sort_values(
    by="importancia",
    ascending=False
))


#NOTE QUE = Segundo o nosso modelo o que é mais importante para definir a posição de um time, seja para ser campeão ou para ser rebaixado,
#é o número de gols marcados. Ou seja, quanto mais gols um time marcar, maiores são as chances de ser campeão. Sendo mas importante
#fazer gols do que sofrer gols, com tanto que ganhe os jogos.

# Os times que foram rebaixados fizeram menos gols e sofreram mais gols, resultando no rebaixamento
```
